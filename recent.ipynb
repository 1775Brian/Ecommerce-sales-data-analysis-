# Cleaning the data

Removing null values

```python
Nan_df=df[df.isna().any(axis=1)]
Nan_df.head()
Nan_df=df.dropna(how='all')
Nan_df.head()
```

Drop Nan

```python
df=df.dropna(how='all')

df.head()
```

# Removing invalid characters

```python
temp_df=df[df['InvoiceDate'].str[0:2]=='1/']
temp_df.head()
df=df[df['InvoiceDate'].str[0:2]!='1/']
df=df[df['InvoiceDate'].str[0:2]!='2/']
df=df[df['InvoiceDate'].str[0:2]!='3/']
df=df[df['InvoiceDate'].str[0:2]!='4/']
df=df[df['InvoiceDate'].str[0:2]!='5/']
df=df[df['InvoiceDate'].str[0:2]!='6/']
df=df[df['InvoiceDate'].str[0:2]!='7/']
df=df[df['InvoiceDate'].str[0:2]!='8/']
df=df[df['InvoiceDate'].str[0:2]!='9/']
```

# Question 4:What products are most often sold together?

```python
df=all_data[all_data['InvoiceNo'].duplicated(keep=False)]

df['Grouped']=df.groupby('InvoiceNo')['Description'].transform(lambda x:','.join(x))
df=df[['InvoiceNo','Grouped']].drop_duplicates()
df.head()
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count=Counter()

for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
for key, value in count.most_common(10):
    print(key,value)
```
