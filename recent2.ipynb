```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/HP/Desktop/Ecommerce/Clean_Data")
all_data=pd.DataFrame(data)
all_data.head()
```

```python
df=all_data[all_data['InvoiceNo'].duplicated(keep=False)]

df['Grouped']=df.groupby('InvoiceNo')['Description'].transform(lambda x:','.join(x))
df=df[['InvoiceNo','Grouped']].drop_duplicates()
df.head()
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count=Counter()

for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
for key, value in count.most_common(10):
    print(key,value)
```
